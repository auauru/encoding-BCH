```python
# BCH 63,45,18
# BCH_encode()  亂數產生 Message 
# BCH_noise()  對 Message 加入高斯雜訊
# BCH_ads()   對 Message 取絕對值


import random
import math
import numpy as np

C=63
M=45
P=C-M
GX=492239

def next_(x):
    x=x<<1;
    if(x&(1<<P)):
        x=x^GX
    return x

def mod(a):
    buf=a>>M;
    for i in range(1,M+1):
        buf=next_(buf) ^ ((a>>(M-i))&1)
    return buf

def BCH_encode():

  m = 24240968276700 #(固定Message)
  #m = int(random.random()*(1<<M)) 
  c=m<<P
  c=c|mod(c);
  BIN=[]

  for i in range(0,C):
    BIN.append((c>>i)&1)
  BIN.reverse()
  print("BCH_encode :",BIN)
  return BIN

def BCH_noise(_message, _r): # _message 訊息 , _r DB值
  BIN=_message
  sigma = math.sqrt((10**-(_r/10))/2)
  for i in range(len(_message)):
    if(BIN[i]==0):
        BIN[i]=1.0+float(np.random.normal(0, sigma, 1))
    else:
        BIN[i]=-1.0+float(np.random.normal(0, sigma, 1))
  #print(sigma)
  print("BCH_noise :",BIN)
  return BIN

def BCH_ads(_message):
  BIN=_message
  for i in range(len(_message)):
    if(BIN[i]<=0):
        BIN[i]=1
    else:
        BIN[i]=0
  return BIN

print("BCH_ads :",BCH_ads(BCH_noise(BCH_encode(),0)))
```
